```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Counts (a) occurrences of a target word in each sentence, and (b) co-occurrence
of 2+ words in each sentence, then aggregates by `part` and makes graphs.

- Word counting is case-insensitive and uses whole-word matching.
- Co-occurrence is computed as:
    * cooccur_flag: 1 if ALL words appear at least once in the sentence, else 0
    * cooccur_min_count: min(count(word_i)) across the words (often 0/1; can be >1)

Outputs:
- sentences_with_counts.csv  (original rows + new count columns)
- by_part_summary.csv        (aggregated by `part`)
- plot_word_by_part.png
- plot_cooccurrence_by_part.png
- plot_word_hist_per_sentence.png
- plot_cooccurrence_hist_per_sentence.png
"""

from __future__ import annotations

import re
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt


# -----------------------------
# CONFIG (edit these)
# -----------------------------
INPUT_CSV = Path(
    "pern_ocr_out/La-ballade-de-Pern-intégrale-McCaffrey_-Anne-La-Ballade-de-Pern_-2015-12-21-9782823821758-bc3525592e.sentences.csv"
)

TEXT_COL = "text"
PART_COL = "part"

TARGET_WORD = "couple"                 # (a)
COOC_WORDS = ["couple", "amour"]       # (b) 2+ words/phrases allowed

OUTPUT_DIR = Path("./outputs_word_counts")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# If you have many parts and want only the top N on plots, set e.g. 20, else None.
TOP_N_PARTS_FOR_PLOTS = None
# -----------------------------


# build a regular expression (regex) that matches a word as a whole word, case-insensitive.
def _whole_word_pattern(term: str) -> re.Pattern:
    """
    Case-insensitive whole-word pattern for a term.
    Works well for single words and short phrases.
    """
    # strip = remove accidental spaces
    # escape = escape special regex characters. e.g., c++ = c\+\+
    term_esc = re.escape(term.strip())
    # whole "word" boundaries around the term
    # (?iu) => ignore case + unicode
    return re.compile(rf"(?iu)\b{term_esc}\b")

# count how many times a word in each sentence
# text_s = one sentence per row (with the panda library)
# term = word to count as str
def count_occurrences_series(text_s: pd.Series, term: str) -> pd.Series:
    """
    Vectorized whole-word counts in a pandas Series of strings.
    """
    pat = _whole_word_pattern(term)
    # pandas uses re internally; passing pattern string is fastest
    # returns integer as output
    return text_s.str.count(pat.pattern, flags=pat.flags).astype("int64")

# returns TRUE/FALSE per sentence if the term is detected
def contains_series(text_s: pd.Series, term: str) -> pd.Series:
    """
    Vectorized whole-word contains for a term.
    """
    pat = _whole_word_pattern(term)
    # na = FALSE to return FALSE is a sentence is missing to avoid crash
    return text_s.str.contains(pat.pattern, flags=pat.flags, na=False)

# used only for plotting
# if have too many books/chapters only return the top
# top_n = None = returns everything
# top_n = 20 keeps only the 20 highest values
def maybe_top_n(df_part: pd.DataFrame, value_col: str, top_n: int | None) -> pd.DataFrame:
    if not top_n:
        return df_part
    return df_part.sort_values(value_col, ascending=False).head(top_n)


def main() -> None:
    # Read (low_memory=False avoids mixed-type warnings on big CSVs)
    df = pd.read_csv(INPUT_CSV, low_memory=False)

    if TEXT_COL not in df.columns:
        raise ValueError(f"Missing column '{TEXT_COL}'. Found: {list(df.columns)}")

    # Normalize part (keep NaN as "Unknown")
    df[PART_COL] = df.get(PART_COL, pd.Series(["Unknown"] * len(df))).fillna("Unknown").astype(str)

    # Ensure text is string (NaN -> "")
    df[TEXT_COL] = df[TEXT_COL].fillna("").astype(str)

    # -----------------------------
    # (a) Count occurrences of TARGET_WORD per sentence
    # -----------------------------
    word_col = f"count__{TARGET_WORD}"
    df[word_col] = count_occurrences_series(df[TEXT_COL], TARGET_WORD)

    # -----------------------------
    # (b) Co-occurrence of COOC_WORDS per sentence
    # -----------------------------
    if len(COOC_WORDS) < 2:
        raise ValueError("COOC_WORDS must contain 2+ items for a co-occurrence measure.")

    # per-word counts (optional but useful)
    cooc_count_cols = []
    for w in COOC_WORDS:
        ccol = f"count__{w}"
        if ccol not in df.columns:
            df[ccol] = count_occurrences_series(df[TEXT_COL], w)
        cooc_count_cols.append(ccol)

    # co-occurrence flag (all present at least once)
    # first put a column of sentence id and a column filled as TRUE
    present_mask = pd.Series(True, index=df.index)
    for w in COOC_WORDS:
        # equals to present_mask = present_mask & contains_series(df[TEXT_COL], w)
        # fill the column above with actual True/False values
        # &= combines the two columns
        # like this first check for couple then for amour
        present_mask &= contains_series(df[TEXT_COL], w)
    
    # returns 1 if the two appear together, 0 if not
    cooc_flag_col = "cooccur__" + "__AND__".join(COOC_WORDS)
    df[cooc_flag_col] = present_mask.astype("int64")

    # co-occurrence min-count (min of individual counts; 0 if any missing)
    # so that couple couple amour amour counts as 2,2 rather than others
    cooc_min_col = "cooccur_min_count__" + "__AND__".join(COOC_WORDS)
    df[cooc_min_col] = df[cooc_count_cols].min(axis=1).astype("int64")

    # -----------------------------
    # Aggregate by PART_COL
    # -----------------------------
    by_part = (
        df.groupby(PART_COL, dropna=False)
          .agg(
              sentences=("sentence_id", "count") if "sentence_id" in df.columns else (TEXT_COL, "count"),
              word_total=(word_col, "sum"),
              word_sentences_with_1plus=(word_col, lambda s: int((s > 0).sum())),
              cooc_sentences=(cooc_flag_col, "sum"),
              cooc_min_count_total=(cooc_min_col, "sum"),
          )
          .reset_index()
    )

    # Save outputs
    out_sentences = OUTPUT_DIR / "sentences_with_counts.csv"
    out_by_part = OUTPUT_DIR / "by_part_summary.csv"
    df.to_csv(out_sentences, index=False)
    by_part.to_csv(out_by_part, index=False)

    # -----------------------------
    # Plots
    # -----------------------------
    # 1) TARGET_WORD totals by part
    plot_df = maybe_top_n(by_part, "word_total", TOP_N_PARTS_FOR_PLOTS)
    plt.figure()
    plt.bar(plot_df[PART_COL], plot_df["word_total"])
    plt.xticks(rotation=45, ha="right")
    plt.ylabel(f"Total occurrences of '{TARGET_WORD}'")
    plt.title(f"Occurrences of '{TARGET_WORD}' by {PART_COL}")
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "plot_word_by_part.png", dpi=200)
    plt.close()

    # 2) Co-occurrence (flag) sentences by part
    plot_df2 = maybe_top_n(by_part, "cooc_sentences", TOP_N_PARTS_FOR_PLOTS)
    plt.figure()
    plt.bar(plot_df2[PART_COL], plot_df2["cooc_sentences"])
    plt.xticks(rotation=45, ha="right")
    plt.ylabel(f"# sentences containing ALL: {COOC_WORDS}")
    plt.title(f"Co-occurrence (ALL present) by {PART_COL}")
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "plot_cooccurrence_by_part.png", dpi=200)
    plt.close()

    # 3) Histogram: per-sentence counts for TARGET_WORD
    plt.figure()
    plt.hist(df[word_col], bins=30)
    plt.xlabel(f"Count of '{TARGET_WORD}' in a sentence")
    plt.ylabel("Number of sentences")
    plt.title(f"Distribution of '{TARGET_WORD}' counts per sentence")
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "plot_word_hist_per_sentence.png", dpi=200)
    plt.close()

    # 4) Histogram: per-sentence cooccurrence min-count
    plt.figure()
    plt.hist(df[cooc_min_col], bins=30)
    plt.xlabel(f"Min-count across {COOC_WORDS} in a sentence")
    plt.ylabel("Number of sentences")
    plt.title("Distribution of co-occurrence min-count per sentence")
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "plot_cooccurrence_hist_per_sentence.png", dpi=200)
    plt.close()

    print("Done.")
    print(f"- Sentences with counts: {out_sentences}")
    print(f"- Summary by part:       {out_by_part}")
    print(f"- Plots in:              {OUTPUT_DIR.resolve()}")


if __name__ == "__main__":
    main()
```
